```python
# import dependencies
import csv 
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
# path to data file
main_file = pd.read_csv("RDC_Inventory_Core_Metrics_State.csv")
main_file.head()
```

```python
# active listings for each state
active_listings = main_file[["state", "active_listing_count"]]
active_listings
```

```python
# line graph of active listings for each state. Too crowded. Need to create groups and plot them separately for increased readability
plt.plot(active_listings['state'], active_listings['active_listing_count'] )
plt.show()
```

```python
# list column names
column_names = list(main_file)
column_names
```

```python
# average listing price year over year
avg_list_y_over_y = main_file[["state", "average_listing_price_yy"]]
avg_list_y_over_y
```

```python
# change in average listing price year over year
# Only Washington DC had a decline in average listing price
avg_list_y_over_y.sort_values(["average_listing_price_yy"])
```

```python
# median listing price year over year
median_list_y_over_y = main_file[["state", "median_listing_price_yy"]]
median_list_y_over_y
```

```python
# median listings year over year sorted. Only DC and West Virginia had the median listing price decrease
median_list_y_over_y.sort_values(["median_listing_price_yy"])
```

```python
new_listings_yy = main_file[["state", "new_listing_count_yy"]]
new_listings_yy.head()
```

```python
# The states with the highest number of new listings, year over year, are in the northeast 
new_listings_yy.sort_values(["new_listing_count_yy"])
```
